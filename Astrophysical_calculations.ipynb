```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Provided data
distance = np.array([88.7144659, 47.98818881, 56.19079122, 26.50219367, 35.57116948,
                     64.89797418, 90.16201494, 56.65837041, 89.48322569, 82.69194642,
                     77.49279864, 25.68075188, 17.94676025, 26.6265791, 61.25172857,
                     24.16249983, 85.62747488, 82.49366569, 75.63146929, 90.12016586])

velocity = np.array([6276.805536, 3379.364009, 3881.284046, 1829.414402, 2496.882592,
                     4525.791526, 6289.84343, 3978.733072, 6403.438857, 5710.467844,
                     5368.381026, 1782.030891, 1271.583113, 1801.607553, 4322.664707,
                     1666.71876, 5988.714796, 5719.472846, 5326.790587, 6379.074542])

# Convert to DataFrame for easier handling
data = pd.DataFrame({'Distance': distance, 'Velocity': velocity})

# Linear Regression (with outliers)
X = data['Distance'].values.reshape(-1, 1)
y = data['Velocity'].values

model = LinearRegression()
model.fit(X, y)

# Hubble's constant with outliers
hubble_constant_with_outliers = model.coef_[0]
intercept_with_outliers = model.intercept_

# Mean and variance (with outliers)
mean_with_outliers = np.mean(y)
variance_with_outliers = np.var(y)

# Identify outliers using the residual method
predicted_velocity = model.predict(X)
residuals = np.abs(y - predicted_velocity)
threshold = 3 * np.std(residuals)  # Outliers: Residuals greater than 3 standard deviations

# Filter out outliers
data_without_outliers = data[residuals <= threshold]

# Linear Regression (without outliers)
X_clean = data_without_outliers['Distance'].values.reshape(-1, 1)
y_clean = data_without_outliers['Velocity'].values

model_clean = LinearRegression()
model_clean.fit(X_clean, y_clean)

# Hubble's constant without outliers
hubble_constant_without_outliers = model_clean.coef_[0]
intercept_without_outliers = model_clean.intercept_

# Mean and variance (without outliers)
mean_without_outliers = np.mean(y_clean)
variance_without_outliers = np.var(y_clean)

# Print results
print("Results with Outliers:")
print(f"Hubble's Constant: {hubble_constant_with_outliers:.2f} km/s/Mpc")
print(f"Mean Velocity: {mean_with_outliers:.2f} km/s")
print(f"Variance of Velocities: {variance_with_outliers:.2f} (km/s)^2\n")

print("Results without Outliers:")
print(f"Hubble's Constant: {hubble_constant_without_outliers:.2f} km/s/Mpc")
print(f"Mean Velocity: {mean_without_outliers:.2f} km/s")
print(f"Variance of Velocities: {variance_without_outliers:.2f} (km/s)^2")
```
